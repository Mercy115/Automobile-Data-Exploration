- Objectives -

Import automobile data set using Python Pandas to answer the following questions.

A. Is there any missing values in the data? If yes, drop the rows with missing values.

B. And answer the following question.
1. which are the cars with minimum or maximum price?
2. How many cars are there with different body-styles?
3. What is the average price of all body types of cars? Arrange them in descending order
4. What is the average price of cars by different companies? Arrange them in descending order
5. Count car types have diffeent number cylinders?

You can also use data visualization to explore the data.


Solution: Colab Notebook uploaded

```python
#A.Checking and dropping the rows with missing values
import pandas as pd

#loading the data
data = pd.read_csv('https://raw.githubusercontent.com/arora123/Data/master/Automobile_data.csv')

# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values
data.dropna(inplace=True)
#Rows with missing values dropped
print(data.isnull().sum())
```

```python
#B_1
#Cars with minimum and maximum prices

#Checking the indices of cars with minimum and maximum prices
min_price_index = data['price'].idxmin()
max_price_index = data['price'].idxmax()

#Getting the correct models of the cars
min_price_car = data.loc[min_price_index, 'company']
max_price_car = data.loc[max_price_index, 'company']

print(f"Car with minimum price: {min_price_car}")
print(f"Car with maximum price: {max_price_car}")
```

```python
#B_2
#Checking no.of cars are with different body styles
num_body_styles = data['body-style'].nunique()
print(f"Number of cars with different body-styles: {num_body_styles}")
```

```python
#B_3
#Arranging the average prices of all body type cars in descending order
avg_price_by_body_style = data.groupby('body-style')['price'].mean().sort_values(ascending=False)
print("Average price of all body types of cars:")
print(avg_price_by_body_style)
```

```python
#B_4
#Arranging the average prices of cars by different companies in descending order
avg_price_by_company = data.groupby('company')['price'].mean().sort_values(ascending=False)
print("Average price of cars by different companies:")
print(avg_price_by_company)
```

```python
#B_5
#Count of car types with different number of cylinders
num_cylinders = data['num-of-cylinders'].nunique()
print(f"Number of car types with different number of cylinders: {num_cylinders}")
```

Exploratory data analysis - Visualization / matplotlib

1. Company & price
2. Company & no of cylinders
3. Body type & price
4. Body type & cylinders
5. engine type & horse power


```python
#Exploratory data analysis-Visualization / matplotlib

import matplotlib.pyplot as plt

#1.Company & price

plt.figure(figsize=(10, 6))
plt.bar(data['company'], data['price'])
plt.xlabel('Company')
plt.ylabel('Price')
plt.title('Company & Price')
plt.xticks(fontsize=5)
plt.show()
```

```python
#2.Company & no of cylinders
plt.figure(figsize=(10, 6))
plt.bar(data['company'], data['num-of-cylinders'])
plt.xlabel('Company')
plt.ylabel('Number of Cylinders')
plt.title('Company & Number of Cylinders')
plt.xticks(fontsize=5)
plt.show()

```

```python
#3.Body type & price
plt.figure(figsize=(10, 6))
plt.bar(data['body-style'], data['price'])
plt.xlabel('Body Type')
plt.ylabel('Price')
plt.title('Body Type & Price')
plt.show()

```

```python
#4.Body type & cylinders
plt.figure(figsize=(10, 6))
plt.bar(data['body-style'], data['num-of-cylinders'])
plt.xlabel('Body Type')
plt.ylabel('Number of Cylinders')
plt.title('Body Type & Number of Cylinders')
plt.show()

```

```python
#5.Engine type & horse power
plt.figure(figsize=(10, 6))
plt.bar(data['engine-type'], data['horsepower'])
plt.xlabel('Engine Type')
plt.ylabel('Horsepower')
plt.title('Engine Type & Horsepower')
plt.show()
```
